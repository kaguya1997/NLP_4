```python
from opencc import OpenCC
from csv import reader,writer
from sklearn import datasets
from sklearn import neighbors
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd
import jieba
import json
import os
```

```python
path = 'wiki_zh'
```

```python
#file_list = os.walk(path)
file_list = os.listdir(path)
```

```python
data = []
for file in file_list:
    print(file)
    path2 = os.path.join(path, file)
    #if os.path.isdir(path2):
    file_list2 = os.listdir(path2)
    for file2 in file_list2:
        print(file2)
        path3 = os.path.join(path2, file2)
        contents = open(path3, encoding="utf-8")
        for content in contents.readlines():
            temp = json.loads(content)
            data.append(temp)
    
    #elif os.path.isfile(path + file):
    #    print(file)
```

```python
len(data)
```

```python
cc = OpenCC('s2t')
```

```python
for i in data:
    i['title'] = cc.convert(i['title'])
    i['title'] = jieba.cut(i['title'])
    i['title'] = [word for word in i['title'] if word != ' ']
    i['title'] = ' '.join(i['title'])
    i['text'] = cc.convert(i['text'])
    i['text'] = jieba.cut(i['text'])
    i['text'] = [word for word in i['text'] if word != ' ']
    i['text'] = ' '.join(i['text'])
```

```python
with open('title_data.txt', 'w', encoding='utf-8') as storage:
    for i in data:
        storage.write(i['title'] + "\n")
```

```python
from gensim.models import word2vec
```

```python
seed = 499
sg = 0
window_size = 20
vector_size = 1000
min_count = 10
workers = 80
epochs = 50
batch_words = 100000

train_data = word2vec.LineSentence('title_data.txt')
model = word2vec.Word2Vec(
    train_data,
    min_count = min_count,
    vector_size = vector_size,
    workers = workers,
    epochs = epochs,
    window = window_size,
    sg = sg,
    seed = seed,
    batch_words = batch_words
)

model.save('title.model')
```

下面是調用'title.model'(即用title所train之model)，且以test_word測試所生成的結果

```python
model = word2vec.Word2Vec.load('title.model')
test_word = '生物'
print(model.wv[test_word].shape)

for item in model.wv.most_similar(test_word, topn=20):
    print(item)
```
